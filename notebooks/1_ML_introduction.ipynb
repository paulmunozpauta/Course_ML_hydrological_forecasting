## ¿Qué es Machine Learning?

Es el arte y la ciencia de programar computadoras para aprender a partir de datos.

> [Arthur Samuel] ML is the field of study that gives computers the ability to learn without being explicitly programmed.

Algo más formal:

> [Tom Mitchell] A computer program is said to learn from experience $E$ with respect to some task $T$ and some performance measure $P$, if its performance on $T$, as measured by $P$, improves with experience $E$.

En hidrología:

> El Aprendizaje Automático (Machine Learning, ML) es una rama de la inteligencia artificial que se centra en el desarrollo de algoritmos y modelos estadísticos que permiten a las computadoras aprender y hacer predicciones o tomar decisiones basadas en datos.

> ML es un subcampo de la inteligencia artificial que desarrolla algoritmos y modelos capaces de extraer patrones y hacer predicciones a partir de datos, sin necesidad de ser programados explícitamente para una tarea.

## ¿Por qué utilizar el Machine Learning?


Cuando construimos sistemas no basados en aprendizaje automático, generalmente seguimos estos pasos:

1. Conceptualización (creamos reglas).
2. Escribimos un algoritmo.
3. Si el algoritmo funciona bien, lo implementamos. Si no, volvemos al paso 1.

Sin embargo, si el problema es complejo, es probable que terminemos con una larga lista de reglas que son difíciles de cumplir y de aplicar a otros problemas similares. Un sistema de aprendizaje automático sería mucho más breve, más fácil de mantener y, en muchos casos, más preciso.

Luego, simplemente podemos entrenar un algoritmo en un gran conjunto de datos, luego inspeccionar la `importancia de características` del algoritmo para obtener una mejor comprensión de la relación entre los datos y el problema. Esto se llama minería de datos.

## Aplicaciones de Machine Learning:


ML tiene muchas aplicaciones, y aquí están algunas de las más destacadas en hidrología y climatologí

- `Clasificación de Imágenes`: generalmente realizada mediante redes neuronales convolucionales.

> En hidrología, esto podría aplicarse a la clasificación de imágenes satelitales para detectar características geográficas y condiciones climáticas.

- `Segmentación Semántica`: el algoritmo se entrena para clasificar cada píxel en una imagen, un ejemplo de esto es la detección de tumores cerebrales.

>En hidrología, se podría aplicar para identificar características geográficas específicas en imágenes satelitales.

- `Procesamiento de Lenguaje Natural` (NLP): más específicamente, la clasificación de texto, que puede aprenderse utilizando RNNs, CNNs o Transformers.

> En aplicaciones de pronóstico del clima, esto podría utilizarse para analizar y clasificar informes meteorológicos y noticias relacionadas con el clima.

- Chatbots: Involucran muchas tareas de NLP, como la Comprensión del Lenguaje Natural (NLU) y la Respuesta a Preguntas.

> En hidrología, podrían utilizarse para general aplicaciones de sistemas de monitoreo y alerta temprana.

- `Pronóstico de Ingresos Futuros`: una tarea de regresión que puede abordarse utilizando múltiples algoritmos como:
    - Regresión Lineal
    - Regresión Polinomial
    - SVM
    - Bosques Aleatorios (Random Forest)
    - Redes Neuronales
    
> En hidrología, esto podría aplicarse al pronóstico de caudales de ríos y niveles de agua.

- `Reconocimiento de Voz`: este problema puede abordarse reconociendo las señales de audio entrantes mediante RNNs, CNNs o Transformers.

> En aplicaciones climáticas, podría utilizarse para transcribir observaciones de campo.

- Detección de Fraude: detectar fraudes puede resolverse mediante aprendizaje supervisado (clasificación) o no supervisado (detección de anomalías)

> En aplicaciones hidrológicas, esto podría utilizarse para detectar anomalías en los datos de sensores de monitoreo de inundaciones.

- `Agrupamiento`: segmentar clientes basados en sus compras para diseñar campañas de marketing dirigidas y más efectivas.

> En hidrología, esto podría aplicarse a la segmentación de regiones geográficas con patrones climáticos similares.

- `Reducción de Dimensionalidad`: útil para la visualización de datos de alta dimensionalidad y el análisis de grupos. Puede resolverse utilizando algoritmos como `PCA` o `T-SNE`.

> En aplicaciones climáticas, esto podría ayudar a visualizar datos de múltiples variables climáticas.

- `Sistemas de Recomendación`: donde podemos alimentar la secuencia de compras de un cliente (por ejemplo) a una red neuronal para predecir la próxima compra.

> En hidrología, esto podría utilizarse para recomendar medidas de mitigación de inundaciones basadas en patrones históricos de eventos de inundaciones.

## Tipos de Sistemas de Aprendizaje Automático

Los algoritmos de ML se pueden clasificar según la cantidad de supervisión que reciben durante el entrenamiento.

Hay 4 tipos principales de algoritmos de ML:

### Aprendizaje Supervisado

Con el aprendizaje supervisado, el conjunto de entrenamiento que alimentamos al algoritmo contiene los objetivos/etiquetas/predicciones deseadas. La mayoría de las tareas de aprendizaje supervisado caen en dos categorías: Clasificación y Regresión.

Con la clasificación, estamos interesados en predecir valores discretos, algunos ejemplos:

- Clasificación de Eventos de Precipitación: Se puede utilizar la clasificación para predecir si un evento de precipitación será lluvia, tormenta o una lluvia intensa. Esto es crucial para la gestión de inundaciones y el monitoreo de eventos climáticos extremos.
- Clasificación de Calidad del Agua: Los algoritmos de clasificación pueden ayudar a determinar si una muestra de agua es potable o no, según los niveles de contaminantes y la presencia de microorganismos dañinos.
- Clasificación de Riesgo de Inundación: Se pueden utilizar modelos de clasificación para evaluar el riesgo de inundaciones en áreas específicas en función de variables como la topografía, el caudal de ríos y la precipitación.
- Clasificación de Tipos de Clima: Los modelos de clasificación pueden agrupar áreas geográficas en función de sus patrones climáticos, como clima tropical, subtropical o árido.


Por otro lado, la regresión se ocupa de valores objetivos continuos, por ejemplo:

- Predicción del Caudal de Ríos: La regresión se utiliza para predecir el caudal de ríos en función de datos históricos de precipitación, temperatura y características geográficas. 
- Modelado de Niveles de Agua en Acuíferos: Los modelos de regresión pueden predecir los niveles de agua en acuíferos subterráneos en función de la recarga, la extracción y otras variables hidrológicas.
- Predicción de Evapotranspiración: La evapotranspiración es un componente clave en el ciclo del agua. La regresión se utiliza para predecir la cantidad de agua que se evapora y se transpira en función de la radiación solar, la temperatura y la humedad.
- Regresión de Temperaturas: La regresión se usa para predecir las temperaturas máximas y mínimas diarias en función de datos meteorológicos históricos, como la radiación solar, la altitud y la latitud.
- Predicción de Precipitación: Se pueden utilizar algoritmos de regresión para predecir la cantidad de precipitación en una región en función de datos climáticos anteriores, como patrones de viento y temperaturas del mar.


Algunos modelos basados en regresión también se utilizan para clasificación, como la `Regresión Logística`, que produce una probabilidad $\in [0,1]$. Aquí tienes algunos algoritmos de aprendizaje supervisado populares:

- K-vecinos más cercanos.
- Regresión Lineal.
- Regresión Logística.
- Árboles de Decisión y Bosques Aleatorios.
- Redes Neuronales.
- Naive Bayes.

### Aprendizaje No Supervisado

En el aprendizaje no supervisado, los datos no tienen etiquetas, y el sistema trata de aprender sin un maestro encontrando estructura interna dentro del conjunto de datos

Algunos algoritmos de aprendizaje no supervisado son:

- Agrupamiento.
  - K-means.

    > K-means para agrupar regiones geográficas en función de patrones climáticos similares. Esto puede ayudar a identificar áreas con climas comparables y facilitar la planificación agrícola y la gestión de recursos hídricos.
  - DBSCAN.
  
    > El algoritmo DBSCAN se puede utilizar para identificar áreas de alta concentración de eventos de lluvia intensa en un conjunto de datos de registros de precipitación. Esto es crucial para identificar zonas propensas a inundaciones.
  
  - Análisis Jerárquico de Clústeres.
  
    > El análisis jerárquico de clústeres se utiliza para organizar regiones geográficas en un árbol de clústeres basado en similitudes climáticas. Esto puede ayudar en la clasificación y comprensión de patrones climáticos a diferentes escalas espaciales.
  
- Detección de Anomalías.
  - One-class SVM.
  
    > One-class SVM para detectar anomalías en series temporales de niveles de agua en ríos. Esto puede ayudar a identificar eventos inusuales como inundaciones repentinas o sequías severas.

  - Isolation Forest.
  
    > El algoritmo Isolation Forest se puede utilizar para identificar puntos de datos atípicos en conjuntos de datos climáticos, lo que podría indicar eventos climáticos extremos, como olas de calor excepcionales o eventos de frío extremo.

  - Autoencoders.
    > Los autoencoders pueden utilizarse para detectar patrones anómalos en registros de calidad del agua. Pueden identificar cambios inusuales en la composición química del agua que podrían indicar problemas de contaminación.
  
- Reducción de Dimensionalidad: el objetivo es comprimir los datos sin perder demasiada información (una forma de hacerlo es combinar características altamente correlacionadas).

  - Análisis de Componentes Principales (PCA).
    > El Análisis de Componentes Principales (PCA) se aplica comúnmente para reducir la dimensionalidad de datos climáticos multivariados. Esto puede simplificar el análisis de patrones climáticos complejos y la visualización de datos.
    
  - Kernel PCA.
  
    > Kernel PCA se puede aplicar para reducir la dimensionalidad de datos climáticos utilizando funciones de kernel no lineales. Esto es útil cuando se trata de datos climáticos altamente no lineales, como patrones de temperatura global.
    
    
  - T-SNE en Hidrología.
    > El Embebimiento Estocástico de Vecinos t-Distribuidos (T-SNE) se utiliza para visualizar datos de series temporales de caudales de ríos en un espacio bidimensional. Esto facilita la identificación de patrones temporales y tendencias en los datos hidrológicos.  
    
    
- Algoritmos de Reglas de Asociación


  - Apriori.
    > Se puede utilizar el algoritmo Apriori para descubrir reglas de asociación interesantes entre variables climáticas, como la relación entre la temperatura y la humedad relativa en diferentes regiones.
    
  - Eclat.

    > Eclat se puede aplicar para encontrar relaciones entre atributos en datos hidrogeológicos, como la asociación entre la permeabilidad del suelo y la recarga de acuíferos en áreas específicas.
    

### Aprendizaje Semi-Supervisado

En el aprendizaje semi-supervisado, tenemos datos parcialmente etiquetados. El objetivo es utilizar datos no etiquetados alrededor de los datos etiquetados como ayuda para resolver la tarea. La mayoría de los algoritmos de aprendizaje semi-supervisado son una combinación de algoritmos de aprendizaje supervisado y no supervisado.


Supongamos que estamos trabajando en un proyecto para predecir inundaciones en una cuenca específica. Tenemos estaciones de monitoreo distribuidas en toda la cuenca que registran datos sobre niveles de agua, precipitación y otros parámetros hidrológicos.

- Datos Etiquetados: Hemos etiquetado algunos eventos pasados de inundaciones y no inundaciones en función de los niveles de agua registrados en varios puntos. Estos datos etiquetados son valiosos, pero representan solo una fracción de todos los datos recopilados.
- Datos No Etiquetados: La mayoría de los datos recopilados no están etiquetados y corresponden a condiciones donde no sabemos si ocurrirá una inundación.
- Modelo de Aprendizaje Supervisado: Comenzamos entrenando un modelo de aprendizaje supervisado utilizando los datos etiquetados para predecir inundaciones. Este modelo inicial tiene un rendimiento razonable pero limitado debido a la falta de datos etiquetados.
- Uso de Datos No Etiquetados: Incorporamos los datos no etiquetados en el proceso de entrenamiento del modelo. A medida que el modelo se expone a más datos no etiquetados, puede aprender patrones y tendencias adicionales en las condiciones previas a las inundaciones.
- Mejora de las Predicciones: A medida que el modelo se entrena con datos no etiquetados, se vuelve más preciso en la predicción de inundaciones, incluso en condiciones que no habíamos encontrado antes.
- Alertas de Inundación en Tiempo Real: Implementamos el modelo mejorado en un sistema de alerta temprana de inundaciones. Ahora, el sistema puede proporcionar alertas en tiempo real basadas en los datos continuos de monitoreo, incluso cuando los datos no están etiquetados.


### Aprendizaje por Reforzamiento

Un agente observa el entorno, selecciona una acción, recibe una recompensa y actualiza su política.
Este enfoque se utiliza en aplicaciones como juegos y control de robots.

> Gestión de Embalses: Imagina un embalse que necesita ser gestionado para garantizar un suministro constante de agua para riego agrícola y consumo humano. Un agente de aprendizaje por refuerzo podría observar las condiciones climáticas actuales, como la cantidad de precipitación y la temperatura, y seleccionar acciones, como liberar o retener agua en el embalse. La recompensa sería la eficiencia en la distribución del agua y la prevención de inundaciones. El agente ajustaría su política a lo largo del tiempo para optimizar la gestión del embalse.

> Predicción de Tormentas: En el contexto de la climatología, un agente de aprendizaje por refuerzo podría observar datos meteorológicos en tiempo real y seleccionar acciones para predecir la intensidad y el impacto de una tormenta en una región específica. La recompensa estaría relacionada con la precisión de la predicción. A medida que el agente recibe datos de tormentas reales, actualizaría su política para mejorar las predicciones futuras.

> Control de Inundaciones: En el ámbito de la hidrología, un agente de aprendizaje por refuerzo podría operar sistemas de compuertas y represas para controlar el flujo de agua en ríos y evitar inundaciones. El agente observaría los niveles de agua actuales y las previsiones meteorológicas y tomaría decisiones para ajustar las compuertas. La recompensa se basaría en la capacidad del agente para mantener los niveles de agua dentro de límites seguros.

> Optimización de Energía Renovable: En el contexto de la energía renovable, un agente de aprendizaje por refuerzo podría gestionar la producción de energía eólica o solar en función de las condiciones climáticas. El agente seleccionaría acciones para maximizar la producción de energía y minimizar los costos. La recompensa estaría relacionada con la eficiencia energética y la rentabilidad.





### Aprendizaje en Lote vs. Aprendizaje en Línea

En el aprendizaje en lote, el modelo es incapaz de aprendizaje incremental y comienza por aprender de todos los datos disponibles sin conexión, luego se despliega para producir predicciones sin alimentarlo con nuevos puntos de datos. Otro nombre para el aprendizaje en lote es Aprendizaje sin Conexión.


Imaginemos que estamos construyendo un modelo de predicción de inundaciones para una cuenca en particular. En el aprendizaje en lote, recopilamos datos históricos de niveles de agua, precipitación y otros parámetros hidrológicos durante muchos años. Luego, utilizamos todos estos datos para entrenar el modelo de predicción de inundaciones de una sola vez. Una vez que el modelo está entrenado, lo desplegamos para que funcione en tiempo real. Durante la operación en tiempo real, el modelo no se actualiza con nuevos datos a medida que llegan. En su lugar, utiliza el conocimiento que adquirió durante el entrenamiento con datos históricos para realizar predicciones de inundaciones.


En el aprendizaje en línea, entrenamos los datos de forma incremental, alimentándolos continuamente a medida que llegan, ya sea individualmente o en pequeños grupos llamados mini-lotes (mini-batches). Cada paso de aprendizaje es rápido y económico, por lo que el sistema puede aprender a medida que llegan los datos, sobre la marcha. El aprendizaje en línea es excelente para sistemas que reciben datos de flujo continuo.

Un aspecto importante del aprendizaje en línea es cuán rápido debe adaptarse el algoritmo de aprendizaje a nuevos puntos de datos o cambios en la distribución general de datos. Con una gran tasa de aprendizaje, el modelo tiende a olvidar datos pasados y se inclina fuertemente hacia nuevos puntos de datos. Con una pequeña tasa de aprendizaje, el modelo tiende a adaptarse ligeramente a nuevos puntos de datos pero mantiene principalmente su conocimiento sobre los datos antiguos.

Un gran desafío con los algoritmos de aprendizaje en línea es que pueden dañarse con malos puntos de datos entrantes y los clientes notarán esto sobre la marcha. Para mitigar esto, podemos monitorear de cerca el sistema a través de métricas de rendimiento y apagar el aprendizaje en línea o volver a un estado de modelo anterior. También debemos asegurarnos de limpiar los datos antes de alimentarlos al modelo mediante detección de anomalías o valores atípicos.

Ahora, consideremos un escenario de pronóstico del tiempo en línea. En lugar de recopilar datos climáticos históricos durante años y entrenar un solo modelo, adoptamos un enfoque de aprendizaje en línea.

En el aprendizaje en línea, continuamente recopilamos nuevos datos climáticos a medida que llegan, como mediciones en tiempo real de temperatura, humedad y presión atmosférica. Estos datos se alimentan en el modelo de pronóstico del tiempo a medida que están disponibles, ya sea individualmente o en pequeños grupos (mini-lotes).

Cada vez que el modelo recibe nuevos datos, ajusta sus predicciones en tiempo real en función de esta información fresca. Esto permite que el pronóstico del tiempo se adapte constantemente a las condiciones climáticas cambiantes y proporcione pronósticos actualizados y precisos.

Sin embargo, es crucial controlar la tasa de aprendizaje en línea para evitar que el modelo reaccione excesivamente a datos ruidosos o anómalos, lo que podría resultar en pronósticos incorrectos. Además, debemos llevar a cabo una detección de anomalías en los datos climáticos para garantizar que solo se utilicen datos confiables para las predicciones.

Mientras que el aprendizaje en lote es adecuado para construir modelos a partir de datos históricos estáticos, el aprendizaje en línea es esencial cuando se trata de sistemas que requieren actualizaciones en tiempo real a medida que llegan nuevos datos, como los sistemas de pronóstico del tiempo.

### Aprendizaje Basado en Instancias vs. Aprendizaje Basado en Modelos

Otra forma de categorizar los algoritmos de aprendizaje automático es cómo generalizan. Hay dos enfoques para la generalización: enfoques basados en instancias y enfoques basados en modelos.

- Con el aprendizaje basado en instancias, realizamos comparaciones basadas en similitud. Un nuevo punto de datos se clasificaría según su similitud con el grupo objetivo en el conjunto de entrenamiento. Esto requeriría una medida de similitud.

Imaginemos que estamos trabajando en un proyecto de monitoreo de calidad del agua en una serie de ríos y arroyos. Tenemos una gran cantidad de datos históricos que incluyen mediciones de parámetros como la temperatura del agua, el nivel de oxígeno disuelto y la concentración de contaminantes en varios puntos a lo largo de estos cuerpos de agua.

En el enfoque de aprendizaje basado en instancias, cuando se recopila un nuevo conjunto de datos que incluye las mismas mediciones, en lugar de construir un modelo que estime directamente la calidad del agua, comparamos estas mediciones con conjuntos de datos históricos similares. Identificamos conjuntos de datos históricos que son más similares en términos de condiciones ambientales, ubicación geográfica, y otros factores, y utilizamos sus resultados de calidad del agua como referencia.

Por ejemplo, si estamos evaluando la calidad del agua en un río en una región montañosa con inviernos fríos, buscamos conjuntos de datos históricos de ríos en regiones similares con características climáticas y geográficas comparables. Luego, usamos los resultados de calidad del agua de esos conjuntos de datos históricos como una indicación de lo que podríamos esperar para el río en cuestión.

- Con el aprendizaje basado en modelos, construimos un modelo para cada clase de puntos de datos y luego usamos el modelo para clasificar un nuevo punto de datos (del entorno de validación/prueba/producción).

Supongamos que estamos desarrollando un modelo de predicción de patrones climáticos en una región específica. En lugar de depender de comparaciones de datos históricos directamente, utilizamos enfoques basados en modelos.

En este caso, construimos modelos estadísticos o de aprendizaje automático para representar patrones climáticos en función de múltiples variables, como la temperatura, la humedad, la presión atmosférica y la topografía. Estos modelos se entrenan con datos históricos para capturar relaciones complejas entre estas variables.

Cuando llega un nuevo conjunto de datos de observaciones climáticas actuales, el modelo basado en modelos utiliza estas observaciones para generar predicciones sobre los patrones climáticos futuros. En lugar de depender únicamente de datos históricos similares, el modelo utiliza su comprensión de las interacciones entre las variables climáticas para hacer pronósticos.

El aprendizaje basado en instancias se basa en comparaciones de similitud con datos históricos similares, mientras que el aprendizaje basado en modelos construye modelos para representar relaciones complejas entre variables y hace predicciones basadas en esos modelos.

### Principales desafíos de ML para pronóstico de caudales y clima

Los dos problemas principales son:

1. Tener datos de mala calidad.
2. Seleccionar un algoritmo de aprendizaje inadecuado.

#### Datos

- Se ha demostrado que muchos algoritmos, desde los más simples hasta los más complejos, tienen un rendimiento relativamente similar cuando se les proporciona suficientes datos.
- SE sugiere entonces mayor esfuerzo en desarrollo de algoritmos de ML o en ingeniería de características (feature engineering).
- El conjunto de datos de entrenamiento también puede no ser representativo del conjunto de datos. Esto represenfra un problema para generalizar de manera efectiva. Si, por ejemplo, el conjunto de entrenamiento es demasiado pequeño, tendremos ruido de muestreo. Incluso muestras grandes pueden no ser representativas si el método de muestreo está defectuoso, esto se llama **sesgo de muestreo**.
- Los datos hidrológicos y climáticos pueden variar significativamente a lo largo del tiempo y el espacio. La selección de la escala adecuada para el modelado es crucial.
- Los datos de baja calidad también pueden ser un gran desafío. La precisión de los datos de precipitación, temperaturas, caudales, y otros parámetros es esencial. Datos inexactos pueden llevar a resultados incorrectos y decisiones erróneas (sobre todo si las mediciones de la variable target son de mala calidad!).
- Si los datos de entrenamiento están llenos de valores atípicos (extremos), errores y ruido, será más difícil para el algoritmo detectar los patrones subyacentes, lo que resultará en un mal modelo. Así mismo, los registros de estaciones meteorológicas o hidrológicas pueden tener valores faltantes o errores de medición. Es necesario abordar estos problemas de manera efectiva.
- Además, los conjuntos de datos pueden contener características (predictores) irrelevantes. Nuestro sistema solo aprenderá si los datos contienen muchas características relevantes y no muchas irrelevantes. Una parte crítica del éxito de un proyecto de aprendizaje automático es lo que se llama **ingeniería de características** o la creación de características que producirían un modelo de calidad.

##### Selección de Características y Feature Engineering

La selección de características y la ingeniería de características (feature engineering) adquieren una importancia adicional:

- **Relevancia Climática e Hidrológica:** Identificar las características climáticas y hidrológicas más relevantes es crucial para modelar correctamente fenómenos como las precipitaciones, caudales de ríos y cambio climático.

- **Inclusión de Variables Específicas:** Variables como la altitud, tipo de suelo, vegetación y características topográficas pueden ser fundamentales en hidrología y climatología.

#### Sobreajuste (Overfitting)

El sobreajuste significa que el modelo de ML funciona bien en los datos de entrenamiento pero falla al generalizar con los datos de prueba.

Modelos complejos como las redes neuronales profundas tienden a memorizar el ruido de los datos de entrenamiento o incluso la muestra de datos en sí si es lo suficientemente pequeña. Podemos hacer lo siguiente para mitigar el sobreajuste:

- Seleccionar un modelo con menos pesos/parámetros para limitar su capacidad predictiva para que solo pueda utilizar los patrones más fuertes presentes.
- Recopilar más datos de entrenamiento.
- Reducir el ruido en los datos de entrenamiento corrigiendo errores y eliminando valores atípicos.

Restringir un modelo y combatir el sobreajuste se llama **regularización**. Si tomamos el ejemplo de un solo modelo de regresión lineal ($f(x)=ax+b$), tiene dos grados de libertad (2 parámetros). Si permitimos que el algoritmo cambie libremente los valores de un parámetro pero tenga un intervalo establecido alrededor del parámetro 2, tendrá entre 1 y 2 grados de libertad. Queremos un buen equilibrio entre mantener el modelo lo más simple posible mientras le damos la capacidad de capturar patrones fuera de los datos de entrenamiento.

La regularización se puede controlar utilizando los hiperparámetros del modelo, que describen cómo debe aprender el modelo. Consideraciones en desarrollo de modelos de caudal y clima:

- **Sobreajuste en Series Temporales:** Es común que los modelos se sobreajusten a las series temporales debido a la complejidad de los patrones climáticos e hidrológicos. Se debe prestar especial atención a la selección de modelos y técnicas de regularización.

- **Incertidumbre Climática y Hidrológica:** Los fenómenos climáticos y hidrológicos están sujetos a incertidumbre inherente. Los modelos deben ser capaces de capturar y cuantificar esta incertidumbre.

#### Subajuste (Underfitting)

El subajuste es lo opuesto al sobreajuste. Ocurre cuando el modelo es demasiado simple para capturar la estructura subyacente de los datos de entrenamiento. Algunas soluciones:
- Podemos seleccionar un modelo más potente, con más parámetros.
- Alimentamos mejores características al algoritmo de aprendizaje (ingeniería de características).
- Reducimos las restricciones en el modelo (reducir la regularización).

#### Transferencia de Conocimiento 

En hidrología y climatología, la transferencia de conocimiento entre regiones o escalas temporales es crucial:

- **Transferencia de Modelos Climáticos:** Adaptar modelos climáticos entrenados en una región a otra puede requerir técnicas avanzadas para ajustar y validar los resultados.

- **Extrapolación Climática:** Prever el comportamiento climático futuro a partir de datos históricos es un desafío importante que requiere técnicas de modelado robustas.

#### Adaptación al Cambio Climático

En el contexto del cambio climático, es fundamental:

- **Incorporación de Escenarios Climáticos Futuros:** Los modelos deben ser capaces de integrar proyecciones climáticas para proporcionar predicciones precisas en un contexto de cambio.

- **Evaluación de Impactos Hidrológicos:** Los modelos deben ser capaces de evaluar los posibles cambios en los patrones de precipitación y caudales, y cómo estos afectarán a los sistemas hidrológicos.

### Prueba y Validación

La prueba y validación de modelos de ML es de suma importancia para asegurar su precisión y fiabilidad en la predicción de fenómenos climáticos e hidrológicos.

Podemos evaluar nuestro modelo dividiendo los datos en dos conjuntos: el conjunto de **entrenamiento** y el conjunto de **prueba**. Solo nos importa el error fuera del entrenamiento, o error de generalización, ya que es representativo del rendimiento del modelo en un entorno de producción.

Si nuestro error de entrenamiento es bajo pero el error de prueba es alto, esto significa que el modelo está sobreajustando. Es común usar el **70%** de los datos para entrenamiento y el 30% restante para pruebas, pero depende del tamaño del conjunto de datos original. Cuanto más grande sea, menos porcentaje podemos tomar como conjunto de pruebas. ¿Y en Hidrología?

Si ajustamos los parámetros de regularización en el conjunto de pruebas, estamos sobreajustando a este conjunto, por lo que necesitamos otro conjunto de datos para la sintonización de hiperparámetros, este conjunto de datos se suele llamar el conjunto de **validación**. El conjunto de validación debe reservarse del conjunto de entrenamiento. Después de realizar la sintonización de hiperparámetros utilizando el conjunto de validación, entrenamos el modelo en el conjunto de entrenamiento completo (con validación) y evaluamos en el conjunto de pruebas.

Una solución simple pero computacionalmente costosa para reservar un conjunto de validación grande es realizar una validación cruzada repetida. Su inconveniente es que tenemos que entrenar el modelo N veces. Tanto el conjunto de validación como el conjunto de pruebas deben ser lo más representativos posible de los datos que utilizaremos en producción. ¿K-fold cross validation?


Algunas consideraciones:

### División de Datos

Para evaluar y validar modelos en hidrología y climatología, es esencial dividir los datos en conjuntos de entrenamiento y prueba de manera cuidadosa. Se deben selecionar **conjuntos representativos** considerando que los patrones climáticos e hidrológicos pueden variar ampliamente según la ubicación y el período de tiempo. Es fundamental asegurarse de que los conjuntos de entrenamiento y prueba sean representativos de las condiciones reales de la región de interés.

### Validación Cruzada

Los datos pueden ser limitados y valiosos, y dada la variabilidad espacial y temporal en datos climáticos e hidrológicos, la validación cruzada puede proporcionar una evaluación más robusta del rendimiento del modelo al abordar diferentes condiciones y ubicaciones.

### Consideraciones sobre el Ruido y los Patrones

Dado que los datos climáticos e hidrológicos pueden estar sujetos a una variedad de fuentes de ruido, como fluctuaciones estacionales o eventos extremos, es importante que los modelos sean capaces de capturar patrones relevantes mientras descartan el ruido. ¿Descomposición de señales?

### Supuestos del Modelo

Los modelos a menudo se basan en supuestos específicos sobre la relación entre las variables. Por ejemplo, un modelo de regresión lineal puede asumir una relación lineal entre las variables, lo que puede no siempre reflejar la complejidad de los fenómenos climáticos e hidrológicos.
